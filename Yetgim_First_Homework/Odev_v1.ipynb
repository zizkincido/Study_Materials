## 1. Kütüphane ve Veri Yükleme
    Pandas ve Numpy kütüphanelerini import edin.
    Size verilen weather_data.csv dosyasını df değişkenine okuyun.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('weather_data.csv')
```

## 2. Veriyi Keşfetme
    Veri setinin ilk 5 satırını ve son 5 satırını ekrana yazdırın.
    Veri setinin istatistiksel özetini görüntüleyin.

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

## 3. Sütun Seçimi
    Sadece Date, City, ve Temperature sütunlarını seçerek yeni bir görünüm oluşturun ve ekrana yazdırın.
    City ve Temperature kolonlarını birlikte listeleyiniz.

```python
df_filtered = df.loc[:, ['Date', 'City', 'Temperature']]

print(df_filtered)
```

```python
print(df_filtered.loc[:, ['City', 'Temperature']])
```

## 4. Basit Filtreleme
    Hava sıcaklığı 30 derecenin üzerinde olan kayıtları listeleyin.
    Şehri sadece 'Bursa' olan kayıtları listeleyin.

```python
print(df[df['Temperature'] > 30])
```

```python
print(df[df['City'] == 'Bursa'])
```

## 5. Mantıksal Operatörler ile Filtreleme
    Şehri 'İstanbul' olan VE nem oranı 60'tan büyük olan kayıtları listeleyin.
    Şehri 'Ankara' olan VEYA sıcaklığı 5 dereceden küçük olan kayıtları listeleyin.
    Sıcaklığı 10 derece VEYA nemi %70 üzerinde olan verileri filtreleyiniz.

```python
print(df[(df['City'] == 'Istanbul') & (df['Humidity'] > 60)])
```

```python
print(df[(df['City'] == 'Ankara') | (df['Temperature'] < 5)])
```

```python
print(df[(df['Temperature'] < 10) | (df['Humidity'] > 70)])
```

## 6. Sıralama (Sorting)
    Veri setini Temperature(Sıcaklık) değerine göre büyükten küçüğe sıralayın ve ilk 10 kaydı gösterin.
    Veriyi neme göre azalan şekilde sıralayınız.
    Veriyi Şehir alanına göre artan şekilde sıralayınız.

```python
df = df.sort_values(by='Temperature', ascending=False)
df.head(10)
```

```python
df = df.sort_values(by='Humidity', ascending=False)
df.head(10)
```

```python
df = df.sort_values(by='City', ascending=True)
df.head(10)
```

## 7. Yeni Sütun Ekleme
    Veri setine 'Temperature_F' adında yeni bir sütun ekleyin. Bu sütun sıcaklığın Fahrenheit cinsinden değerini taşımalıdır.
    'Feelslike' adında yeni bir kolon oluşturun. Formül: Temperature - (Humidity / 100)

```python
df['Temperature_F'] = (df['Temperature'] * 9/5) + 32
df['FeelsLike'] = df['Temperature'] - (df['Humidity'] / 100)
df.head()
```

## 8. Gruplama(Grouping) ve Analiz:
    Her şehirde kaç adet veri kaydı olduğunu hesaplayın.
    Şehirlere göre ortalama sıcaklık değerlerini hesaplayın.

```python
count_city = df['City'].groupby(df['City']).count()
print(count_city)
```

```python
avg_temp = df['Temperature'].groupby(df['City']).mean()
print(avg_temp)
```

## 9. En Yüksek/Düşük Değer Analizi
    En yüksek sıcaklığın olduğu satırı(tüm bilgilerini) bulun ve yazdırın.
    En düşük nem oranının olduğu satırı bulun ve yazdırın.

```python
max_temp = df[df['Temperature'] == df['Temperature'].max()]
print(max_temp)
```

```python
min_humidity = df[df['Humidity'] == df['Humidity'].min()]
print(min_humidity)
```

## 10. Dışa Aktarma(Export)
    8. maddede bulduğunuz 'Şehirlere göre ortalama sıcaklık' tablosunu bir Excel veya CSV dosyası olarak kaydedin.

```python
avg_temp.to_csv('sehir_sicaklikları.csv')
```
